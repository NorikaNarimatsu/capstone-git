```python
from crossref.restful import Works
import pandas as pd
from tqdm import tqdm

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Create panda dataframe for storing results
df = pd.DataFrame(columns=['year', 'month', 'volume', 'page', 'title', 'URL', 'authors', 'journal'])

# Use ISSN to filter 
works = Works()
r = works.filter(issn='0748-4658')  

for item in r:
    # If there is no item, replaece by None
    if ('author' not in item):
        item['author'] = None
    if ('page' not in item):
        item['page'] = None
    if ('volume' not in item):
        item['volume'] = 0
    # This jounal has two data (publication/print and publication/online)
    if ('published-print' in item):
        year = item['published-print']['date-parts'][0][0]
        month = item['published-print']['date-parts'][0][1]
    elif ('published-online' in item):
        year = item['published-online']['date-parts'][0][0]
        month = item['published-online']['date-parts'][0][1]
    temp_se = pd.Series([year, month, item['volume'], item['page'],
                        item['title'][0], item['URL'], item['author'],
                        item['container-title'][0]], index = df.columns)
    # add rows by appending elements to df
    df = df.append(temp_se, ignore_index = True)  


#pbar.close()
# csvで保存する
df = df.sort_values(['year', 'month', 'URL'])  # 順番バラバラなのでソートする
df
```
